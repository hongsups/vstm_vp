```python
from scipy.stats import gamma
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
"""
    Define a function to compute MATLAB equivalent of gampdf 
        with proper x values that covers p from 0.00001 to 0.99999
    Input: Jbar (float!)
    par_scale: scale parameter (defaul value = 1.0)
    Output:
        x_vec: x vector
        y: gampdf 
"""
def gampdf(Jbar,par_scale=1.0):
    par_shape = Jbar/par_scale # scipy.stats.gamma needs an input of shape parameter
    x_vec = np.linspace(gamma.ppf(0.00001,par_shape),gamma.ppf(0.99999,par_shape),100)
    y = gamma.pdf(x_vec,par_shape)
    return x_vec, y
```

```python
# draw plot
fig, ax = plt.subplots(1, 1)

# Change font setting
font = {'family' : 'Arial',
        'weight' : 'bold',
        'size'   : 20}
import matplotlib
matplotlib.rc('font', **font)

# Enlarge the figure
fig.set_size_inches(18.5, 10.5)

# Assign colors
ax.set_color_cycle(['c', 'm', 'y', 'k'])

# loop over set sizes to plot gampdf
N_vec = [x*2 for x in range(1,5)] # set size vector
for N in N_vec:
    power = 1
    Jbar = 10.0 * N ** (-power)
    x_vec, y = gampdf(Jbar)
    ax.plot(x_vec, y, lw = 2, label='N='+str(N))
    
plt.xlabel('J')
plt.ylabel('probability') # precision
plt.legend(loc='best', frameon=False)
plt.show()    
```
